```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/mnt/d/piicture/pexels-fox-1172675.jpg")
B, G, R = cv2.split(img)

print(img.shape)
print(f"Shape of Blue channel: {B.shape}")
print(f"Shape of Green channel: {G.shape}")
print(f"Shape of Red channel: {R.shape}")
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1); plt.imshow(B, cmap="gray"); plt.title("Blue Channel")
plt.subplot(1, 4, 2); plt.imshow(G, cmap="gray"); plt.title("Green Channel")
plt.subplot(1, 4, 3); plt.imshow(R, cmap="gray"); plt.title("Red Channel")

plt.subplot(1, 4, 4); plt.imshow(img[..., ::-1]); plt.title("Original image(RGB)")
```

```python
merged = cv2.merge((B, G, R))

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1); plt.imshow(B, cmap="gray"); plt.title("Blue Channel")
plt.subplot(1, 4, 2); plt.imshow(G, cmap="gray"); plt.title("Green Channel")
plt.subplot(1, 4, 3); plt.imshow(R, cmap="gray"); plt.title("Red Channel")

plt.subplot(1, 4, 4); plt.imshow(merged[..., ::-1]); plt.title("Merged image(RGB)")
```

```python
zero = np.zeros_like(R, 'uint8')

blue = cv2.merge((B, zero, zero))
green = cv2.merge((zero, G, zero))
red = cv2.merge((zero, zero, R))

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1); plt.imshow(blue, cmap="gray"); plt.title("Blue Channel")
plt.subplot(1, 4, 2); plt.imshow(green, cmap="gray"); plt.title("Green Channel")
plt.subplot(1, 4, 3); plt.imshow(red, cmap="gray"); plt.title("Red Channel")

plt.subplot(1, 4, 4); plt.imshow(merged[..., ::-1]); plt.title("Merged image(RGB)")
```

```python
import cv2
```

```python
img_path = "/mnt/d/piicture/pexels-fox-1172675.jpg"

img = cv2.imread(img_path)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2.imshow("HSV Image", hsv_img)
cv2.imshow("Hue channel", hsv_img[:, :, 0])
cv2.imshow("Saturation channel",  hsv_img[:, :, 1])
cv2.imshow("Value channel", hsv_img[:, :, 2])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 3))

plt.subplot(1, 3, 1); plt.imshow(hsv_img[:, :, 0], cmap='gray')
plt.subplot(1, 3, 2); plt.imshow(hsv_img[:, :, 1], cmap='gray')
plt.subplot(1, 3, 3); plt.imshow(hsv_img[:, :, 2], cmap="gray")
```

```python
saturated_img = cv2.merge([hsv_img[:, :, 0], cv2.add(hsv_img[:, :, 1], 100), hsv_img[:, :, 2]])
saturated_img = cv2.cvtColor(saturated_img, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1); plt.imshow(saturated_img); plt.title("Saturated Image")
plt.subplot(1, 2, 2); plt.imshow(img[..., ::-1]); plt.title("Original Image")
```

```python
brighter_img = cv2.merge([hsv_img[:, :, 0], hsv_img[:, :, 1], cv2.add(hsv_img[:, :, 2], 50)])
brighter_img = cv2.cvtColor(brighter_img, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1); plt.imshow(brighter_img); plt.title("Brighter Image")
plt.subplot(1, 2, 2); plt.imshow(img[..., ::-1]); plt.title("Original Image")
```
