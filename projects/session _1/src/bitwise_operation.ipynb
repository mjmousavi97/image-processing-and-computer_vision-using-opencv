```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
square = np.zeros(shape=(300, 300), dtype=np.uint8)
ellipse = np.zeros(shape=(300, 300), dtype=np.uint8)

cv2.rectangle(square, (50, 50), (250, 250), 255, -1)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)

plt.subplot(1, 2, 1); plt.imshow(square, cmap="gray"); plt.title("square");plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(ellipse, cmap="gray"); plt.title("ellipse"); plt.axis("off")
```

```python
bitwise_and = cv2.bitwise_and(square, ellipse)
bitwise_or = cv2.bitwise_or(square, ellipse)
bitwise_xor = cv2.bitwise_xor(square, ellipse)
bitwise_not_square = cv2.bitwise_not(square)
bitwise_not_ellipse = cv2.bitwise_not(ellipse)

plt.figure(figsize=(20, 4))
plt.subplot(1, 5, 1); plt.imshow(bitwise_and, cmap='gray'); plt.title("And")
plt.subplot(1, 5, 2); plt.imshow(bitwise_or, cmap='gray'); plt.title("Or")
plt.subplot(1, 5, 3); plt.imshow(bitwise_xor, cmap='gray'); plt.title("Xor")
plt.subplot(1, 5, 4); plt.imshow(bitwise_not_square, cmap='gray'); plt.title("Not square")
plt.subplot(1, 5, 5); plt.imshow(bitwise_not_ellipse, cmap='gray'); plt.title("Not Ellipse");
```

```python
img = cv2.imread("/mnt/d/piicture/pexels-fox-1172675.jpg")
img_cropped = img[1750:2250, 2800:3300]
img_cropped = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2RGB)

plt.imshow(img_cropped)
```

```python
circle = np.zeros(shape=(500, 500, 3), dtype=np.uint8)
cv2.circle(circle, (250, 250), 250, (255, 255, 255), -1)

output_img = cv2.bitwise_and(img_cropped, circle)

plt.imshow(output_img)
```

```python
img_png = np.zeros((500, 500, 4), np.uint8)
circle_one_channel = np.zeros((500, 500), np.uint8)
cv2.circle(circle_one_channel, (250, 250), 250, 250, -1)

img_png[..., 0:3] = output_img
img_png[..., 3] = circle_one_channel

plt.imshow(img_png)
img_png = cv2.cvtColor(img_png, cv2.COLOR_RGB2BGR)
cv2.imwrite("/mnt/e/Image processing and computer vision using OpenCV/outputs/output1.png", img_png)
```
