```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
image_bw = np.zeros(shape=(512, 512), dtype=np.uint8)
```

```python
plt.figure(figsize=(10, 5))
           
plt.subplot(1, 2, 1); plt.imshow(image)
plt.subplot(1, 2, 2);plt.imshow(image_bw, cmap='gray')
```

```python
cv2.line(image, (50, 0), (511, 511), (255, 255, 0), 5);
plt.imshow(image[..., ::-1])
```

```python
cv2.rectangle(image, (100, 100), (300, 250), (0, 0, 255), 3)
cv2.imshow("Rectangle", image)

cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(image[..., ::-1])
```

```python
cv2.rectangle(image, (100, 100), (300, 250), (0, 0, 255), -3)
cv2.imshow("Rectangle", image)

cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(image[..., ::-1])
```
