```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8)

cv2.putText(img, "Hello world!", (150, 256), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 125, 0), 3)
cv2.imshow("Hello World", img)

cv2.waitKey()
cv2.destroyAllWindows()

plt.imshow(img[..., ::-1])
```

```python
import arabic_reshaper
from bidi.algorithm import get_display
from PIL import Image, ImageDraw, ImageFont
```

```python
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
```

```python
text = "سلام دنیا"

reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)
text = bidi_text.encode().decode(encoding="utf-8")

pil_image = Image.fromarray(img)
font = ImageFont.truetype("arial", 100, encoding="unic")
draw = ImageDraw.Draw(pil_image)
draw.text((120, 200), text, (255, 0, 0), font=font)

plt.imshow(pil_image)

```
