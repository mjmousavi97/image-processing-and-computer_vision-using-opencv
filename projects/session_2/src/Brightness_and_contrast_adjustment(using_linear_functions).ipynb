```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# g(x) = alpha*image + beta, alpha > 0
image = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/input.jpg")

alpha_1 = float(input("Enter alpha_1: "))
beta_1 = int(input("Enter beta_1: "))
alpha_2 = float(input("Enter alpha_2: "))
beta_2 = int(input("Enter beta_2: "))

new_image_1 = cv2.convertScaleAbs(image, alpha=alpha_1, beta=beta_1)
new_image_2 = cv2.convertScaleAbs(image, alpha=alpha_2, beta=beta_2)

plt.figure(figsize=(24, 8))
plt.subplot(1, 3, 1); plt.imshow(image[..., ::-1]); plt.title("Original Image")
plt.subplot(1, 3, 2); plt.imshow(new_image_1[..., ::-1]); plt.title(f"Scaled Image with alpha = {alpha_1} and beta = {beta_1}")
plt.subplot(1, 3, 3); plt.imshow(new_image_1[..., ::-1]); plt.title(f"Scaled Image with alpha = {alpha_2} and beta = {beta_2}")
```

## Histogram Analysis

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_1_gray = cv2.cvtColor(new_image_1, cv2.COLOR_BGR2GRAY)
image_2_gray = cv2.cvtColor(new_image_2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(24, 8))
plt.subplot(1, 3, 1); plt.hist(image_gray.ravel(), 256, (0, 255)); plt.title("Histogram of gray image")
plt.subplot(1, 3, 2); plt.hist(image_1_gray.ravel(), 256, (0, 255)); plt.title("Histogram of gray image_1")
plt.subplot(1, 3, 3); plt.hist(image_2_gray.ravel(), 256, (0, 255)); plt.title("Histogram of gray image_2")
```
