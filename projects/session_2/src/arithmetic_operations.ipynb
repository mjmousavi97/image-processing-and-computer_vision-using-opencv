```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/input.jpg"
image = cv2.imread(image_path)

m = np.ones(shape=image.shape, dtype=np.uint8) * 100

print(f"Type of m is: {m.dtype}")
print(f"Type of image is: {image.dtype}")

# new_image = image + m
add_image = cv2.add(image, m)
sub_image = cv2.subtract(image, m)

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1); plt.imshow(add_image[..., ::-1]); plt.title("Image + m")
plt.subplot(1 ,2, 2); plt.imshow(sub_image[..., ::-1]); plt.title("Image - m")
```

```python
image_vision = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/class.vision.jpg")
shape = image_vision.shape
image_input = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/input.jpg")[100:100+shape[0], 100:100+shape[1], :]
print(f"Shape of image vision: {image_vision.shape}")
print(f"shape of input image: {image_input.shape}")

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1); plt.imshow(image_vision[..., ::-1]); plt.title("Class Vision")
plt.subplot(1, 2, 2); plt.imshow(image_input[..., ::-1]); plt.title("Cropped image input")
```

```python
output_image = cv2.add(image_input, image_vision)

plt.imshow(output_image[..., ::-1])
```

```python
image_1 = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/input.jpg")
image_2 = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/class.vision.jpg")

print(f"Shape of image vision: {image_2.shape}")
print(f"shape of input image: {image_1.shape}")

shape_1 = image_1.shape
shape_2 = image_2.shape
```

```python
padded_iamge = np.zeros_like(image_1)

padded_iamge[10:10+shape_2[0], 500:500+shape_2[1], :] = image_2
print(f"Shape of padded image : {padded_iamge.shape}")
```

```python
output_image = cv2.add(image_1, padded_iamge)

plt.imshow(output_image[..., ::-1])
```

```python
weighted_add_imagee = cv2.addWeighted(image_1, 1, padded_iamge, 5, 10)  # alpha*image1 + beta*image2 + gamma

plt.imshow(weighted_add_imagee[..., ::-1])
```
