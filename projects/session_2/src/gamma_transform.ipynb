```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/room-light.jpg")
cv2.imshow("Original Image", image)

plt.imshow(image[..., ::-1]); plt.title("Original Image"); plt.axis("off")

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
lookup = np.zeros(shape=(1, 256), dtype=np.uint8) 
gamma = 0.4

for i in range(256):
    lookup[0][i] = np.clip((pow((i / 255), gamma)) * 255, 0, 255)
    print(f"Intensity {i} --> {lookup[0][i]}")
```

```python
correct_image = cv2.LUT(image, lut = lookup)

plt.imshow(correct_image[..., ::-1])
```
