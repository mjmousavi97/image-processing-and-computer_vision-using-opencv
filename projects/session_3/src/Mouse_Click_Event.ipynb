## Brush

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
events = [i for i in dir(cv2) if "EVENT" in i]
events
```

```python
image = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
LBDOWN = False
color = (255, 0, 0)

def brush(event, x, y, flag, param):
    global image
    global color
    global LBDOWN
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(image, (x, y), 5, color, -1)
        LBDOWN = True

    elif event == cv2.EVENT_MOUSEMOVE:
        if LBDOWN == True:
            cv2.circle(image, (x, y), 5, color, -1)

    elif event == cv2.EVENT_LBUTTONUP:
        LBDOWN = False

cv2.namedWindow("Image")
cv2.setMouseCallback("Image", brush)

while True:
    cv2.imshow("Image", image)
    
    key = cv2.waitKey(1)

    if (key & 0xFF) == 27:
        break
    elif (key & 0xFF) == ord("g"):
        color = (0, 255, 0)
    elif (key & 0xFF) == ord("r"):
        color = (0, 0, 255)
    elif (key & 0xFF) == ord("b"):
        color = (255, 0, 0)

cv2.destroyAllWindows()

plt.imshow(image[..., ::-1])
```

# Paint

```python
image = np.zeros((512, 512, 3), dtype=np.uint8)
left_button = False
ix = 0
iy = 0

def paint(event,x, y, flag, param):
    global left_button 
    global ix, iy

    if event == cv2.EVENT_LBUTTONDOWN:
        left_button = True
        ix = x
        iy = y

    elif event == cv2.EVENT_MOUSEMOVE:
        if left_button == True:
            cv2.line(image, (ix, iy), (x, y), (255, 255, 255), 2)
            ix = x
            iy = y

    elif event == cv2.EVENT_LBUTTONUP:
        left_button = False

cv2.namedWindow("Image")
cv2.setMouseCallback("Image", paint)

while True:
    cv2.imshow("Image", image)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()

plt.imshow(image[..., ::-1])
```

## Rectangle

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = np.zeros((512, 512, 3), dtype=np.uint8)
left_button = False
ix = 0
iy = 0

def paint(event,x, y, flag, param):
    global left_button 
    global ix, iy

    if event == cv2.EVENT_LBUTTONDOWN:
        left_button = True
        ix = x
        iy = y

    elif event == cv2.EVENT_MOUSEMOVE:
        if left_button == True:
            cv2.rectangle(image, (ix, iy), (x, y), (255, 255, 255), -1)

    elif event == cv2.EVENT_LBUTTONUP:
        left_button = False

cv2.namedWindow("Image")
cv2.setMouseCallback("Image", paint)

while True:
    cv2.imshow("Image", image)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()

plt.imshow(image[..., ::-1])
```
