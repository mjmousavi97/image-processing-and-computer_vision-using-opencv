```python
import cv2
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/class.vision.jpg")
image2 = cv2.imread("/mnt/e/Image processing and computer vision using OpenCV/class.vision-master/images/opencv.jpg")

def add_tracker(val):
    global added_image, alpha

    alpha = val / 100
    beta = 1.0 - alpha
    added_image = cv2.addWeighted(image1, alpha, image2, beta, 0.0)
    cv2.imshow("Image Blend", added_image)


cv2.namedWindow("Image Blend")
cv2.createTrackbar("Track", "Image Blend", 0, 100, add_tracker)

add_tracker(0)

cv2.waitKey()
cv2.destroyAllWindows()

plt.imshow(added_image[..., ::-1]); plt.title(f"Added image1 and image2 with tracker valu: {alpha}");
```

```python
import numpy as np
```

```python
image = np.zeros((512, 512, 3), dtype=np.uint8)

def do_nothing(valu):
    pass

cv2.namedWindow("Image RGB")
cv2.createTrackbar("RED", "Image RGB", 0, 255, do_nothing)
cv2.createTrackbar("GREEN", "Image RGB", 0, 255, do_nothing)
cv2.createTrackbar("BLUE", "Image RGB", 0, 255, do_nothing)

r = cv2.getTrackbarPos("RED", "Image RGB")
g = cv2.getTrackbarPos("GREEN", "Image RGB")
b = cv2.getTrackbarPos("BLUE", "Image RGB")

while True: 
    r = cv2.getTrackbarPos("RED", "Image RGB")
    g = cv2.getTrackbarPos("GREEN", "Image RGB")
    b = cv2.getTrackbarPos("BLUE", "Image RGB")

    image[:, :, 0] = b
    image[:, :, 1] = g
    image[:, :, 2] = r

    cv2.imshow("Image RGB", image)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()

plt.imshow(image[..., ::-1]); plt.title(f"Image with value: R = {r}, G = {g} and B = {b}");
```
